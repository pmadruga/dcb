```python
import pandas as pd
```

```python
df = pd.read_csv('../data/jobindex_2021.csv')
```

```python
# df.dtypes
df['date'] = pd.to_datetime(df['date'])
df['title'] = df['title'].astype(str)
df.dtypes
```

```python
df['title'].describe()
# df.groupby('title').agg(['count', 'size', 'nunique']).stack()
```

```python
# df[~df['title'].str.contains("no_title", na=False)]
```

```python
mydf1 = (df.set_index('date')
               .groupby([pd.Grouper(freq = 'Y'), 'title'])
               .size().reset_index())
mydf1['title'].describe()
```

```python
mydf1.columns = ['date', 'title', 'count']
print(mydf1.columns)
```

```python
# mydf1.sort_values(by='title',ascending=True).head(50)
df_before_2020 = mydf1[(mydf1['date'] >= '2017-01-01') & (mydf1['date'] <= '2019-12-31')].sort_values('title')
```

```python
df_before_2020.describe()
```

```python
df_before_2020_average = (df_before_2020.set_index('date')
#                .groupby([pd.Grouper(freq = 'Y'), 'title'])
                . groupby('title')
               .mean().reset_index())
```

```python
df_before_2020_average.head(50)
```

```python
# mydf1.sort_values(by='title',ascending=True).head(50)
df_2020 = mydf1[(mydf1['date'] >= '2020-01-01') & (mydf1['date'] <= '2020-12-31')].sort_values('title')
```

```python
df_2020.describe()
```

```python
 merge_res = pd.merge(df_before_2020_average, df_2020, left_on='title', right_on='title')
```

```python
merge_res.describe()
```

```python
df_variation = merge_res[['title', 'count_x', 'count_y']]
```

```python
df_variation['count_x'] = df_variation['count_x'].astype(int)
```

```python
df_variation.columns = ['title', 'avg_before_2020', '2020']
df_variation
```

```python
# s = pd.Series([90, 85])
df_calculated_variation = df_variation[['avg_before_2020', '2020']].T.pct_change().T

df_calculated_variation
```

```python
df_changes = pd.DataFrame()

# df_changes[] = df_variation[['avg_before_2020', '2020']]
df_changes['title'] = df_variation['title']
df_changes['avg_before_2020'] =  df_variation['avg_before_2020']
df_changes['2020_total'] =  df_variation['2020']
df_changes['change(%)'] = df_calculated_variation['2020'].apply(lambda x: x*100)
```

```python
df_changes.head()
```

```python
df_decline = df_changes[df_changes['avg_before_2020'] > 50].sort_values('change(%)', ascending=True).head(50)
df_growth = df_changes[df_changes['avg_before_2020'] > 50].sort_values('change(%)', ascending=False).head(50)
```

```python
# df_growth['percentage_change'] = df_growth['percentage_change'].apply(lambda x: "{:.2%}".format(x))

df_growth.to_csv('growth.csv')
df_decline.to_csv('decline.csv')
```
